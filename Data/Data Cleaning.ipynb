```python
delta = gp.read_file('delta.geojson')
ogun = gp.read_file('ogun.geojson')
kaduna = gp.read_file('kaduna.geojson')
lagos = gp.read_file('lagos.geojson')
```

```python
delta = delta.drop(['id', 'geometry_type','global_id','category','education','poi_type','number_of_students',
                      'number_of_teachers','ward_code','source','lga_code','state_code','geometry'], axis =1)
ogun = ogun.drop(['id', 'geometry_type','global_id','category','education','poi_type','number_of_students',
                      'number_of_teachers','ward_code','source','lga_code','state_code','geometry'], axis =1)
kaduna = kaduna.drop(['id', 'geometry_type','global_id','category','education','poi_type','number_of_students',
                      'number_of_teachers','ward_code','source','lga_code','state_code','geometry'], axis =1)
lagos = lagos.drop(['id', 'geometry_type','global_id','category','education','poi_type','number_of_students',
                      'number_of_teachers','ward_code','source','lga_code','state_code','geometry'], axis =1)
```
